## Test YOLO on Datasets

```python
results = {}
results['Real'] = yolo.test_real()
results['Augmented'] = yolo.test_augmented()
results['Real (mosaic)'] = yolo.test_real_mosaic()
results['Augmented (mosaic)'] = yolo.test_augmented_mosaic()
results['Real (mixup)'] = yolo.test_real_mixup()
results['Augmented (mixup)'] = yolo.test_augmented_mixup()
```

## Comparative Table of Performance Metrics

```python
# Define the metrics you want to extract
metrics = ['metrics/precision(B)', 'metrics/recall(B)', 'metrics/mAP50(B)', 'metrics/mAP50-95(B)', 'fitness']

# Use a loop to construct the data dictionary for DataFrame
data = {'Case': []}
for metric in metrics:
    data[metric] = []

# Fill the data dictionary
for case_name, result in results.items():
    data['Case'].append(case_name.capitalize())
    for metric in metrics:
        data[metric].append(result.results_dict[metric])

# Calculate additional metrics F1 Score and Accuracy
f1_scores = []
accuracies = []  # Example placeholder, since accuracy typically requires true negatives which aren't provided here

for i in range(len(data['Case'])):
    precision = data['metrics/precision(B)'][i]
    recall = data['metrics/recall(B)'][i]
    if precision + recall != 0:
        f1_score = 2 * (precision * recall) / (precision + recall)
    else:
        f1_score = 0
    f1_scores.append(f1_score)

# Adding the new metrics to the data dictionary
data['F1 Score'] = f1_scores

# Creating the DataFrame
df = pd.DataFrame(data)

# Renaming the columns for better readability
df.columns = ['Case', 'Precision', 'Recall', 'mAP50', 'mAP50-95', 'Fitness', 'F1 Score']

print(df)

# Export to LaTeX
latex_code = df.to_latex(index=False, float_format="{:0.3f}".format)
print(latex_code)
```

## Confusion Matrices

```python
# Adjusting global font sizes
plt.rcParams['font.size'] = 16         # Default font size
plt.rcParams['axes.titlesize'] = 20  # Axes title font size
plt.rcParams['axes.labelsize'] = 18  # X and Y axis label font size
plt.rcParams['xtick.labelsize'] = 14  # X axis tick labels
plt.rcParams['ytick.labelsize'] = 14 # Y axis tick labels
plt.rcParams['legend.fontsize'] = 18   # Legend font size

def plot_confusion_matrix(matrix_data, class_names, normalize=False, show_numbers=False, ax=None, xlabel = True, ylabel = True):
    
    if ax is None:
        fig, ax = plt.subplots(figsize=(10, 8))
    
    # Append 'background' to the class names if not enough names
    #class_labels = [class_names.get(i, 'background') for i in range(len(matrix_data))]
    class_labels = [i if i in class_names.keys() else np.nan for i in range(len(matrix_data))]
    # Convert to numpy array for manipulation
    matrix = np.array(matrix_data)
    if normalize:
        # Normalize the columns
        column_sums = np.sum(matrix, axis=0)
        matrix = matrix / column_sums[np.newaxis, :]
    
    cax = ax.matshow(matrix, cmap=cm.Blues)
    #plt.colorbar(cax, ax=ax)
    
    # Set ticks and labels
    ax.set_xticks(np.arange(len(class_labels)))
    ax.set_yticks(np.arange(len(class_labels)))
    ax.set_xticklabels(class_labels, rotation=30)
    ax.set_yticklabels(class_labels)
    
    # Text annotations inside the squares
    for i in range(len(matrix)):
        for j in range(len(matrix[0])):
            if normalize:
                text = f'{matrix[i, j]:.2f}'
                if show_numbers:
                    text += f'\n ({int(matrix_data[i][j])})'
            else:
                text = f'{int(matrix[i, j])}'
            ax.text(j, i, text, va='center', ha='center', 
                    color='black' if matrix[i, j] < 0.5 * matrix.max() else 'white')
    
    if xlabel: ax.set_xlabel('True labels')
    if ylabel: ax.set_ylabel('Predicted labels')
    #ax.set_title('Confusion Matrix' + (' (Normalized)' if normalize else ''))

# Create a figure and a grid of subplots
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(16, 24))
# Flatten the axes array for easier indexing
axes = axes.flatten()

i = 0
for case_name, result in results.items():
    ax = axes[i]
    conf_matrix = result.confusion_matrix
    matrix_data = conf_matrix.matrix
    class_names = result.names
    plot_confusion_matrix(matrix_data,
                          class_names,
                          normalize=True,
                          show_numbers=True,
                          ax=ax,
                          xlabel = (i>1),
                          ylabel = not (i%2))
    ax.set_title(f'{case_name}')
    i+=1
    
# Show the plot
plt.tight_layout()
os.makedirs("./figs/fig4", exist_ok=True)
plt.savefig("./figs/fig4/ablation_confmat.pdf", format="pdf", bbox_inches='tight', dpi=300)
```
